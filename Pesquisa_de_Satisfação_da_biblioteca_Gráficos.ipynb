#**Pesquisa de Satisfação do Cine Biblioteca Maria Dolores**



Este projeto teve como objetivo analisar os dados coletados em uma pesquisa de satisfação realizada no Cine Biblioteca, que funciona na Biblioteca Comunitária Maria Dolores. A pesquisa ocorreu no período de 2024 a 2025, contando com a participação de 152 pessoas da comunidade local.



**Etapas do Projeto**

**Importação das Bibliotecas:**

Para manipulação e visualização dos dados, foram importadas as bibliotecas Python Pandas e Matplotlib.

Pandas foi utilizada para organizar os dados em um DataFrame, facilitando a manipulação e análise.

Matplotlib foi a ferramenta escolhida para a criação dos gráficos que ilustram os resultados da pesquisa de forma clara e objetiva.


**Organização dos Dados:**

Os dados da pesquisa foram estruturados em um DataFrame chamado frequencia, com as seguintes colunas:

Resposta: Categorias das respostas dos entrevistados ("sim", "não", "não responderam").

Você já frequentou um cinema?: Quantidade de respostas para essa pergunta.

Você recomendaria para seus amigos ou familiares?: Quantidade de respostas para essa outra pergunta.

Esse formato permite a comparação visual entre as duas questões levantadas pela pesquisa.

**Visualização dos Dados:**

Para tornar os dados acessíveis e compreensíveis para o público, foi criado um gráfico de barras lado a lado usando o Matplotlib.

Cada grupo de barras representa as respostas para cada categoria (sim, não, não responderam).

O gráfico mostra claramente quantas pessoas já frequentaram o cinema e quantas recomendariam o Cine Biblioteca para amigos ou familiares.

Valores numéricos são exibidos no topo de cada barra para facilitar a leitura rápida dos resultados.



**Resultados e Observações:**

Observou-se que a maioria dos participantes já frequentou o cinema (124 pessoas) e que um número ainda maior (140 pessoas) recomendaria o espaço para amigos ou familiares, evidenciando uma boa aceitação do Cine Biblioteca na comunidade.

Apenas um pequeno grupo não frequentou o cinema ou não recomendaria, e poucos não responderam às perguntas.

## **Gráficos**

### **1. Você já frequentou um cinema?**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados
frequencia = pd.DataFrame({
    "Resposta": ["sim", "não", "não responderam"],
    "Você já frequentou um cinema?": [124, 20, 9]
})

valores = frequencia["Você já frequentou um cinema?"]
labels = frequencia["Resposta"]

# Cores
cores = ['#4285F4', '#EA4335', '#FBBC05']

# Função para formatar os textos da legenda
def gerar_legendas(valores, labels):
    total = sum(valores)
    legendas = [
        f"{label} - {valor/total:.1%}".replace('.', ',')  # pt-BR
        for label, valor in zip(labels, valores)
    ]
    return legendas

# Criando o gráfico
fig, ax = plt.subplots(figsize=(7, 7))
wedges, texts = ax.pie(
    valores,
    labels=None,
    colors=cores,
    startangle=90,
    wedgeprops={'edgecolor': 'white'}
)

# Título
ax.set_title("Você já frequentou um cinema?", fontsize=14)

# Adiciona legenda com os rótulos formatados
legenda_formatada = gerar_legendas(valores, labels)
ax.legend(wedges, legenda_formatada, title="Respostas", loc="center left", bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()

```

### **2. Você recomendaria para seus amigos ou familiares?**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados
frequencia = pd.DataFrame({
    "Resposta": ["sim", "não", "não responderam"],
    "Você recomendaria para seus amigos ou familiares?": [140, 4, 9]
})

valores = frequencia["Você recomendaria para seus amigos ou familiares?"]
labels = frequencia["Resposta"]

# Cores
cores = ['#4285F4', '#EA4335', '#FBBC05']  # azul, vermelho, amarelo

# Função para criar legendas personalizadas
def gerar_legendas(valores, labels):
    total = sum(valores)
    legendas = [
        f"{label} - {valor/total:.1%}".replace('.', ',')
        for label, valor in zip(labels, valores)
    ]
    return legendas

# Criação do gráfico
fig, ax = plt.subplots(figsize=(7, 7))

wedges, _ = ax.pie(
    valores,
    labels=None,
    colors=cores,
    startangle=90,
    wedgeprops={'edgecolor': 'white'}
)

# Título
ax.set_title("Você recomendaria para seus amigos ou familiares?", fontsize=13)

# Adiciona legenda com os rótulos formatados
legenda_formatada = gerar_legendas(valores, labels)
ax.legend(wedges, legenda_formatada, title="Respostas", loc="center left", bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```

### **3. Atendimento/recepção dos voluntários**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados da coluna específica
dados = {
    "Avaliação": ["Muito satisfeito", "Satisfeito", "Regular", "Insatisfeito", "Muito insatisfeito", "Não Respondeu"],
    "Atendimento/recepção dos voluntários": [112, 34, 2, 1, 2, 1]
}

df = pd.DataFrame(dados)
valores = df["Atendimento/recepção dos voluntários"]
labels = df["Avaliação"]

cores = ['#4CAF50', '#8BC34A', '#FFEB3B', '#FFC107', '#FF5722', '#9E9E9E']  # cores agradáveis

def gerar_legenda_formatada(valores, labels):
    total = sum(valores)
    return [f"{label} - {valor/total:.1%}".replace('.', ',') for label, valor in zip(labels, valores)]

# Gráfico
fig, ax = plt.subplots(figsize=(7, 7))
wedges, _ = ax.pie(
    valores,
    labels=None,
    colors=cores,
    startangle=90,
    wedgeprops={'edgecolor': 'white'}
)

ax.set_title("Atendimento/recepção dos voluntários", fontsize=13)
legenda = gerar_legenda_formatada(valores, labels)
ax.legend(wedges, legenda, title="Avaliações", loc="center left", bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```

### **4. Atendimento/recepção dos voluntários**

```python
import matplotlib.pyplot as plt

# Função para gerar a legenda formatada
def gerar_legenda_formatada(valores, labels):
    total = sum(valores)
    return [f"{label} - {valor/total:.1%}".replace('.', ',') for label, valor in zip(labels, valores)]

# Função principal para plotar o gráfico de pizza
def plotar_grafico_pizza(titulo, valores):
    labels = ["Muito satisfeito", "Satisfeito", "Regular", "Insatisfeito", "Muito insatisfeito", "Não Respondeu"]
    cores = ['#4CAF50', '#8BC34A', '#FFEB3B', '#FFC107', '#FF5722', '#9E9E9E']  # Cores por nível de satisfação

    fig, ax = plt.subplots(figsize=(7, 7))
    wedges, _ = ax.pie(
        valores,
        labels=None,
        colors=cores,
        startangle=90,
        wedgeprops={'edgecolor': 'white'}
    )

    ax.set_title(titulo, fontsize=13)
    legenda = gerar_legenda_formatada(valores, labels)
    ax.legend(wedges, legenda, title="Avaliações", loc="center left", bbox_to_anchor=(1, 0.5))

    plt.tight_layout()
    plt.show()
```

```python
# Exemplo: Atendimento/recepção dos voluntários
plotar_grafico_pizza(
    "Atendimento/recepção dos voluntários",
    [112, 34, 2, 1, 2, 1]
)
```
